## Random seed
Set the random seed to a certain value for reproducibility.

```
import random

import torch
import numpy as np
import torch
from torchvision.utils import save_image
!pip3 install stylegan2_pytorch
from stylegan2_pytorch import ModelLoader

def same_seeds(seed):
    # Python built-in random module
    random.seed(seed)
    # Numpy
    np.random.seed(seed)
    # Torch
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.benchmark = False
    torch.backends.cudnn.deterministic = True

same_seeds(2021)
```

```
workspace_dir = '/content/drive/MyDrive/Course_ML2021Spring/HW06'


loader = ModelLoader(
    base_dir = workspace_dir+'/faces',   # path to where you invoked the command line tool
    name = 'default'                   # the project name, defaults to 'default'
)

noise   = torch.randn(1, 512).cuda() # noise
styles  = loader.noise_to_styles(noise, trunc_psi = 0.1)  # pass through mapping network
images  = loader.styles_to_images(styles) # call the generator on intermediate style vectors

save_image(images, workspace_dir+'/sample.jpg') # save your images, or do whatever you desire
```
